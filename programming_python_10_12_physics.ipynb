```
import matplotlib.pyplot as plt
v = 10 #初速度 v0 [m/s]
x = 0.0 # 初期位置 x0 [m]
 
t = 0.0 # 時間 t[s]
dt = 0.01 # 区切る時間の幅Δt 

TIME_END = 20

tlist = []
xlist = []

while int(t) < TIME_END:
  t += dt
  x += v*dt
  # [x に，Δt の間に進んだ距離を足す]

  tlist.append(t)
  xlist.append(x)

plt.plot(tlist, xlist)
plt.show()
```

```
import matplotlib.pyplot as plt
g = 9.8 # 重力加速度 g[m/s^2]
v = 10 #初速度 v0 [m/s]
x = 100 # 初期高度 x0 [m]
 
t = 0.0 # 時間 t[s]
dt = 0.01 # 区切る時間の幅Δt 

TIME_END = 20

tlist = []
xlist = []

while x > 0:
  t += dt
  v += g * dt
  # [vに，Δt の間に増えた速度を足す]
  x -= v * dt
  # [x から，Δt の間に落ちた距離を引く]

  tlist.append(t)
  xlist.append(x)

plt.plot(tlist, xlist)
plt.show()
```

```
# -*- coding: utf-8 -*-
# 9/17 用演習プログラム ロケットの逆噴射
import numpy as np
import matplotlib.pyplot as plt

# 定数   
g = 9.8 

# tf 秒後に逆噴射の加速度を返す関数

def retrofire(t, tf):
    if t >= tf:
        return -1.5 * g  #重力の1.5倍のS加速度で逆噴射を行う（ロケットの重さは考えない）
    else:
        return 0.0;    

t = 0.0   # 時刻t
dt = 0.01  # 時刻の刻み幅

# 係数の入力
v = float(input("初速度v0を入力してください:"))
x0 = float(input("初期高度x0を入力してください:"))
tf = float(input("開始時刻tfを入力してください:"))
x = x0  # 初期高度の設定


tlist = [t]
xlist = [x]
# 自由落下の計算
while (x > 0) and (x <= x0):  
    t += dt                           # 時刻の更新
    v += (g + retrofire(t, tf)) * dt                                 # 速度の計算
    x -= v * dt                                 # 位置の更新

    tlist.append(t)
    xlist.append(x)
# グラフの表示
plt.plot(tlist, xlist)  # グラフをプロット
plt.show()


```

```
# バネの振動

import matplotlib.pyplot as plt
m = 3 # 重りの重さ[kg]
x = 1 # 重りの初期位置[m]
k = 5 # バネ係数 [N/m]
x0 = 0 # バネの自然長の位置を原点とする。 [m]
t = 0.0 # 時間 t[s]
dt = 0.01 # 区切る時間の幅Δt
v = 0 # 初速度 [m/s] 重りは静止した状態から始めてよい

TIME_END = 20

tlist = []
xlist = []

while t < 10:
  t += dt
  a = -1 * k/m * (x-x0)
  v += a * dt
  x += v * dt

  tlist.append(t)
  xlist.append(x)

plt.plot(tlist, xlist)
plt.xlabel("t [s]")
plt.ylabel("x [m]")
plt.show()


import matplotlib.pyplot as plt
m = 3 # 重りの重さ[kg]
x = 1 # 重りの初期位置[m]
k = 5 # バネ係数 [N/m]
x0 = 0 # バネの自然長の位置を原点とする。 [m]
t = 0.0 # 時間 t[s]
dt = 0.01 # 区切る時間の幅Δt
v = 0 # 初速度 [m/s] 重りは静止した状態から始めてよい

TIME_END = 20

tlist = []
xlist = []

while t < 10:
  t += dt
  a = -1 * k/m * (x-x0)
  v += a * dt
  x += v * dt

  tlist.append(t)
  xlist.append(x)

plt.plot(tlist, xlist)
plt.xlabel("t [s]")
plt.ylabel("x [m]")
plt.show()
```

```
# 地面との間に摩擦が生じる場合(動摩擦力)

import matplotlib.pyplot as plt
import numpy as np # ヒント: np.sign()を使用して、運動の向きを取り出す
m = 3 # 重りの重さ[kg]
x = 1 # 重りの初期位置[m]
k = 5 # バネ係数 [N/m]
x0 = 0 # バネの自然長の位置を原点とする。 [m]
t = 0.0 # 時間 t[s]
dt = 0.01 # 区切る時間の幅Δt
v = 0 # 初速度 [m/s] 重りは静止した状態から始めてよい
g = 9.8 #[m/s^2]
ud = 0.01 # μ' の値

TIME_END = 20

tlist = []
xlist = []

while t < TIME_END:
  t += dt
  a = -1 * k/m * (x-x0) - ud * g * np.sign(v)
  v += a * dt
  x += v * dt

  tlist.append(t)
  xlist.append(x)

plt.plot(tlist, xlist)
plt.xlabel("t [s]")
plt.ylabel("x [m]")
plt.show()
```

```
# 重りが水中を移動する場合（速度に比例した粘性抵抗が働くとき）

import matplotlib.pyplot as plt
import numpy as np # ヒント: np.sign()を使用して、運動の向きを取り出す
m = 3 # 重りの重さ[kg]
x = 1 # 重りの初期位置[m]
k = 5 # バネ係数 [N/m]
x0 = 0 # バネの自然長の位置を原点とする。 [m]
t = 0.0 # 時間 t[s]
dt = 0.01 # 区切る時間の幅Δt
v = 0 # 初速度 [m/s] 重りは静止した状態から始めてよい

alpha = 0.01 # 液体の粘性抵抗

TIME_END = 2000

tlist = []
xlist = []

while t < TIME_END:
  t += dt
  a = -1 * k/m * (x-x0) - alpha/m * v
  v += a * dt
  x += v * dt

  tlist.append(t)
  xlist.append(x)

plt.plot(tlist, xlist)
plt.xlabel("t [s]")
plt.ylabel("x [m]")
plt.show()
```

```
# -*- coding: utf-8 -*-
# 数式による再現と，オイラー法によるシミュレーション結果 【応用】
# モジュールのインポート
import numpy as np
import matplotlib.pyplot as plt

# 定数
F = 1.5     
G = 9.80665  

# 下請け関数の定義
# retrofire()関数
def retrofire(t, tf):
    if t >= tf:
        return -F * G  
    else:
        return 0.0;    
# retrofire()関数の終わり

# メイン実行部
t = 0.0   # 時刻t
h = 0.01  # 時刻の刻み幅

# 係数の入力
v = float(input("初速度v0を入力してください:"))
x0 = float(input("初期高度x0を入力してください:"))
tf = float(input("開始時刻tfを入力してください:"))
x = x0  # 初期高度の設定
print("{:.7f} {:.7f} {:.7f}".format(t, x, v))  # 現在時刻と現在の位置
# グラフデータに現在位置を追加
tlist = [t]
xlist = [x]
# 自由落下の計算
while (x > 0) and (x <= x0):  
    t += h                           # 時刻の更新
    v += (G + retrofire(t, tf)) * h  # 速度の計算
    x -= v * h                       # 位置の更新
    if round(t,2) == round(tf,2) :
      
      v1 = -v
      x1 = x
      print("逆噴射開始判定 t:", t, "v1:", v1, "x1:", x1)

    #print("{:.7f} {:.7f} {:.7f}".format(t, x, v))  # 現在時刻と現在の位置
    # グラフデータに現在位置を追加
    tlist.append(t)
    xlist.append(x)
# グラフの表示
plt.plot(tlist, xlist)  # グラフをプロット
#plt.ylim([0, 200])
plt.show()
# glander.pyの終わり

c1 = v1 - G*(F-1)*tf
c2 = x1 - 0.5*G*(F-1)*tf*tf - c1*tf
print("c1:", c1, "c2:", c2)

x_b = []
x_a = []
for tl in tlist:
  x_b.append(-1 * 0.5*G*tl*tl + x0)
  x_a.append(0.5*G*(F-1)*tl*tl + c1*tl + c2)

plt.plot(tlist, x_b)
plt.plot(tlist, x_a)
plt.ylim([0, 100])
plt.show()




```
